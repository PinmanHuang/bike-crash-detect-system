```python
from __future__ import print_function

# Import MNIST data
# from tensorflow.examples.tutorials.mnist import input_data
# mnist = input_data.read_data_sets("/tmp/data/", one_hot=True)

import tensorflow as tf
import numpy as np
```

## File Path ##

```python
TRAIN_PATH = "./result2/train.csv"
TEST_PATH = "./test2.csv"
TEST_PATH_1 = "./result2/train.csv"
MODEL_PATH = "./model/"
```

## Load Data ##

```python
training_set = tf.contrib.learn.datasets.base.load_csv_with_header(
    filename=TRAIN_PATH,
    target_dtype=np.int,
    features_dtype=np.float32)

''''''
test_set = tf.contrib.learn.datasets.base.load_csv_with_header(
    filename=TEST_PATH,
    target_dtype=np.int,
    features_dtype=np.float32)

test_set_1 = tf.contrib.learn.datasets.base.load_csv_with_header(
    filename=TEST_PATH_1,
    target_dtype=np.int,
    features_dtype=np.float32)
''''''

feature_columns = [tf.feature_column.numeric_column("x", shape=[50])]
```

## Classifier ##

```python
classifier = tf.estimator.DNNClassifier(
    feature_columns=feature_columns,
    hidden_units=[20, 50, 50, 20, 10],
    n_classes=2,
    model_dir=MODEL_PATH)
```

## Train ##

```python
train_input_fn = tf.estimator.inputs.numpy_input_fn(
    x={"x": np.array(training_set.data)},
    y=np.array(training_set.target),
    batch_size=1,
    num_epochs=None,
    shuffle=True)
```

```python
classifier.train(input_fn=train_input_fn, steps=100)
```

## Test ##

```python
test_input_fn = tf.estimator.inputs.numpy_input_fn(
    x={"x": np.array(test_set.data)},
    y=np.array(test_set.target),
    batch_size=1,
    num_epochs=50,
    shuffle=False)
accuracy_score = classifier.evaluate(input_fn=test_input_fn)["accuracy"]
print("nTest Accuracy: {0:f}n".format(accuracy_score))
```

```python
test_input_fn_1 = tf.estimator.inputs.numpy_input_fn(
    x={"x": np.array(test_set_1.data)},
    y=np.array(test_set_1.target),
    batch_size=1,
    num_epochs=1,
    shuffle=False)
accuracy_score = classifier.evaluate(input_fn=test_input_fn_1)["accuracy"]
print("nTest Accuracy: {0:f}n".format(accuracy_score))
```

## Predict ##

```python
predictions = list(classifier.predict(input_fn=test_input_fn))
predicted_classes = [p["classes"] for p in predictions]

print("New Samples, Class Predictions:    {}\n".format(predicted_classes))
```
